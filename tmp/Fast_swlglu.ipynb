```python
import numpy as np
import matplotlib.pyplot as plt

# Fast SwiGLU Activation Function
def fast_swiglu(x):
    return x * (np.maximum(0, x + 1) - np.maximum(0, x - 1))

# Generate a range of values from -3 to 3
x = np.linspace(-3, 3, 300)
# Apply the fast_swiglu function to these values
y = fast_swiglu(x)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="Fast SwiGLU")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Fast SwiGLU Activation Function")
plt.legend()
plt.grid(True)
plt.show()

```
