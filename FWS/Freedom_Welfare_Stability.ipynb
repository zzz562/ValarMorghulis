# FWS Equation

## Variables
- **Freedom (F)**: Represented as a percentage of leisure time in a year or inversely proportional to the working hours.
- **Welfare (W)**: Expressed as a multiple of the average social income, with 1.0 as the baseline.
- **Stability (S)**: A derivative of the current state of freedom and welfare over time.

## Equation
Given:
- \( F(t) \): Freedom as a function of time.
- \( W(t) \): Welfare as a function of time.
- \( $\alpha$, $\beta$ \): Coefficients to adjust the impact of changes in freedom and welfare on stability.

The stability can be calculated as:
$\[ S(t) = -\alpha \left( \frac{dF(t)}{dt} \right)^2 - \beta \left( \frac{dW(t)}{dt} \right)^2 \]$

Where $\( \frac{dF(t)}{dt} \) $ and $ \( \frac{dW(t)}{dt} \) $ are the rates of change of freedom and welfare over time, respectively.


```python
# 
import matplotlib.pyplot as plt
import numpy as np

# 10
t = np.linspace(0, 10, 1000)  # 101000

# FW
F = 0.1 * t  # 
W = 1.0 + 0.5 * np.sin(np.pi * t)  # 

# S
alpha, beta = 0.1, 0.1  # 
dF_dt = np.gradient(F, t)  # F
dW_dt = np.gradient(W, t)  # W
S = -alpha * dF_dt**2 - beta * dW_dt**2  # 

# 
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(t, F)
plt.title("Freedom (F) Over Time")
plt.xlabel("Time (Years)")
plt.ylabel("Freedom")

plt.subplot(1, 3, 2)
plt.plot(t, W)
plt.title("Welfare (W) Over Time")
plt.xlabel("Time (Years)")
plt.ylabel("Welfare")

plt.subplot(1, 3, 3)
plt.plot(t, S)
plt.title("Stability (S) Over Time")
plt.xlabel("Time (Years)")
plt.ylabel("Stability")

plt.tight_layout()
plt.show()
```
