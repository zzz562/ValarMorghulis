```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 创建一个网格
freedom = np.linspace(0, 1, 100)
welfare = np.linspace(0, 1, 100)
freedom, welfare = np.meshgrid(freedom, welfare)

# 应用不可能三角的约束条件
# Freedom + Welfare + Stability <= 2
stability = 2 - freedom - welfare
stability[stability < 0] = np.nan  # 移除不符合条件的值

# 创建图表
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 绘制曲面
surf = ax.plot_surface(freedom, welfare, stability, cmap='viridis')

# 设置坐标轴
ax.set_xlabel('Freedom')
ax.set_ylabel('Welfare')
ax.set_zlabel('Stability')

# 设置标题
ax.set_title('The Impossible Triangle of Freedom, Welfare, and Stability')

# 显示图形
plt.show()

# 可选：保存图表为文件
fig.savefig('3D_Surface_Plot.png')  # 保存图表到本地路径

from IPython.display import Image
Image(filename='3D_Surface_Plot.png')
```

# FWS Equation

## Variables
- **Freedom (F)**: Represented as a percentage of leisure time in a year or inversely proportional to the working hours.
- **Welfare (W)**: Expressed as a multiple of the average social income, with 1.0 as the baseline.
- **Stability (S)**: A derivative of the current state of freedom and welfare over time.

## Equation
Given:
- \( F(t) \): Freedom as a function of time.
- \( W(t) \): Welfare as a function of time.
- \( $\alpha$, $\beta$ \): Coefficients to adjust the impact of changes in freedom and welfare on stability.

The stability can be calculated as:
$\[ S(t) = -\alpha \left( \frac{dF(t)}{dt} \right)^2 - \beta \left( \frac{dW(t)}{dt} \right)^2 \]$

Where $\( \frac{dF(t)}{dt} \) $ and $ \( \frac{dW(t)}{dt} \) $ are the rates of change of freedom and welfare over time, respectively.


```python
# 导入所需库
import matplotlib.pyplot as plt
import numpy as np

# 定义时间范围（例如10年）
t = np.linspace(0, 10, 1000)  # 10年，1000个时间点

# 定义自由度（F）和福利（W）的函数
F = 0.1 * t  # 线性增长的自由度
W = 1.0 + 0.5 * np.sin(np.pi * t)  # 周期性变化的福利

# 计算稳定性（S）
alpha, beta = 0.1, 0.1  # 系数
dF_dt = np.gradient(F, t)  # F对时间的导数
dW_dt = np.gradient(W, t)  # W对时间的导数
S = -alpha * dF_dt**2 - beta * dW_dt**2  # 计算稳定性

# 绘制图表
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(t, F)
plt.title("Freedom (F) Over Time")
plt.xlabel("Time (Years)")
plt.ylabel("Freedom")

plt.subplot(1, 3, 2)
plt.plot(t, W)
plt.title("Welfare (W) Over Time")
plt.xlabel("Time (Years)")
plt.ylabel("Welfare")

plt.subplot(1, 3, 3)
plt.plot(t, S)
plt.title("Stability (S) Over Time")
plt.xlabel("Time (Years)")
plt.ylabel("Stability")

plt.tight_layout()
plt.show()
```
